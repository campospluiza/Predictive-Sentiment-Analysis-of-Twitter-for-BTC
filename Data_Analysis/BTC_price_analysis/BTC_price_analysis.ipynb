```python
fields = ['Date','Opening','Closing']
BTC_daily_movement = pd.read_csv('Bases de Datos\Historical_price_btc_2016-2023.csv',usecols = fields)
BTC_daily_movement.head()
```

```python
BTC_daily_movement['Date'] = pd.to_datetime(BTC_daily_movement['Date'])
```

```python
print(BTC_daily_movement['Date'].dtype)
```

```python
print(BTC_daily_movement)
```

```python
BTC_daily_movement['Opening'] = BTC_daily_movement['Opening'].str.replace(',', '')
BTC_daily_movement['Opening'] = BTC_daily_movement['Opening'].astype(float)
```

```python
BTC_daily_movement['Difference'] = BTC_daily_movement['Opening'] - BTC_daily_movement['Closing']
BTC_daily_movement.head()
```

```python
BTC_daily_movement.insert(2, 'Closing', BTC_daily_movement.pop('Closing'))
BTC_daily_movement.head()
```

```python
BTC_daily_movement.info()
```

```python
opening_price = BTC_daily_movement.Opening.values
closing_price = BTC_daily_movement.Closing.values 
```

```python
fig = plt.figure()
ax = plt.axes()

ax.scatter(opening_price,closing_price)
ax.set(xlabel='Opening', ylabel= 'Closing', title='Opening_price vs Closing_price')
```

```python
date = BTC_daily_movement.Date.values
difference_price = BTC_daily_movement.Difference.values 
```

```python
fig = plt.figure()
ax = plt.axes()

ax.scatter(date,difference_price)
ax.set(xlabel='Date_year', ylabel= 'Difference_price', title='Difference_price per year')
```
