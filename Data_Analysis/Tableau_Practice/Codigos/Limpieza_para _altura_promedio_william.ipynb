```python
#creamos una función para convertir los datos
def str_a_float(dato):
    dato = dato.replace(" ","") #eliminamos los espacios sobrantes
    dato = dato.replace("-","'") # camniamos el guion por apostrofe
    dato = dato.replace("''","") # eliminamos el doble apostofe
    dato = dato.replace('"','') # eliminamos las dobles comillas
    dato = dato.replace("'",".")
    #dato = dato[:-1] # seleccionamos todos los elementos del str menos el último
    dato = float(dato) #convertimos el str a
    return dato
```

```python
#aplicamos la función que creamos a la columna del data frame
df_altura_promedio["height_clean"] = df_altura_promedio["height_clean"].apply(lambda x: str_a_float(x))
```
