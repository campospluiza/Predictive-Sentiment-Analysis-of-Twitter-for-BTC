```python
import tweepy
import pandas as pd
from textblob import TextBlob
import numpy as np
```

```python
# Tokens de acceso
consumer_key = ''
consumer_secret = ''
access_token = ''
access_token_secret = ''
```

```python
# Autenticaci√≥n en la API de Twitter
auth = tweepy.OAuth1UserHandler(consumer_key, consumer_secret, access_token, access_token_secret)
api = tweepy.API(auth)
```

```python
# Define las palabras clave o cuentas de Twitter relevantes
palabras_clave = []
cant_palabras_clave = int(input('Cuantas Palabras clave va a ingresar? '))
for i in range(1, cant_palabras_clave + 1):
    input_palabras_clave = input(f'Ingresa las palabras clave #{i}: ')
    palabras_clave.append(input_palabras_clave)

cuentas_twitter = ['@elonmusk', '@CoinDesk'] # Ejemplo de cuentas de Twitter
print(palabras_clave)
```

```python
# Obt√©n tweets basados en las palabras clave o cuentas de Twitter
tweets = []
for palabra_clave in palabras_clave:
    for tweet in tweepy.Cursor(api.search_tweets, palabra_clave, count=100, tweet_mode='extended').items(500):
        tweets.append(tweet)

for cuenta in cuentas_twitter:
    for tweet in tweepy.Cursor(api.user_timeline, screen_name=cuenta, count=100, tweet_mode='extended').items(500):
        for palabra_clave in palabras_clave:
            if palabra_clave in tweet.full_text.lower():
                tweets.append(tweet)
                break
```

```python
# Almacena los tweets en un DataFrame de Pandas
tweets_df = pd.DataFrame(columns=['Fecha', 'Usuario', 'Texto', 'Sentimiento', 'Interacciones'])
for tweet in tweets:
    fecha = tweet.created_at
    usuario = tweet.user.screen_name
    texto = tweet.full_text
    sentimiento = TextBlob(texto).sentiment.polarity
    interacciones = tweet.retweet_count + tweet.favorite_count
    tweet_df = pd.DataFrame({'Fecha': [fecha], 'Usuario': [usuario], 'Texto': [texto], 'Sentimiento': [sentimiento], 'Interacciones': [interacciones]})
    # Concatenar el DataFrame del tweet con el DataFrame principal
    tweets_df = pd.concat([tweets_df, tweet_df], ignore_index=True)

# Ordenar los tweets por fecha y resetear el √≠ndice
tweets_df = tweets_df.sort_values(by='Fecha').reset_index(drop=True)

# Imprimir el DataFrame
tweets_df.head(50)
```
